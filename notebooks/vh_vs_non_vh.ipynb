```python
import numpy as np
from matplotlib.patches import Patch
import matplotlib.pyplot as plt
```

```python
vh_back = dict(
    fin=tuple((-0.8297, -0.847)),
    hun=tuple((-0.0915, -2.1996)),
    tur=tuple((-3.1502, -4.0729)),
    mnc=tuple((-2.5563, -3.4993)),
)

nvh_back = dict(
    arb=tuple((-0.4452, -0.0874)),
    ain=tuple((-0.8201, 0.808)),
    ekk=tuple((0.0984, -0.1489)),
    hye=tuple((0.3118, -0.3299)),
    eus=tuple((-0.246, 0.2473))
)
```

```python
vh_round = dict(
    khk=tuple((-1.8211, -0.6621)),
    tur=tuple((-1.0516, -3.185)),
)

nvh_round = dict(
    arb=tuple((-0.0677, 0.3892)),
    ain=tuple((-0.808, 0.8201)),
    ekk=tuple((0.094, -0.1394)),
    hye=tuple((-0.251, -0.3369)),
    eus=tuple((-0.246, 0.2447))
)
```

```python
fig = plt.figure(figsize=(16, 7))
gs = fig.add_gridspec(1, 2, hspace=0.15, wspace=0.25)
(ax_back, ax_round) = gs.subplots(sharex=True,sharey=True)

for lang_id, (delta_mu_front, delta_mu_back) in vh_back.items():
    ax_back.scatter(delta_mu_front, delta_mu_back, color='red')
    ax_back.text(x=delta_mu_front+0.1, y=delta_mu_back+0.1, s=lang_id, fontsize=14)
for lang_id, (delta_mu_front, delta_mu_back) in nvh_back.items():
    ax_back.scatter(delta_mu_front, delta_mu_back, color='green')
    ax_back.text(x=delta_mu_front+0.04, y=delta_mu_back, s=lang_id, fontsize=14)
ax_back.grid(True, which='both')
ax_back.set_xlabel("delta_mean_front", fontsize=20)
ax_back.set_ylabel("delta_mean_back", fontsize=20)
ax_back.tick_params(labelsize = 15, axis='both')

ax_back.axhline(y = 0.0, color = 'black', linestyle = '-')
ax_back.axvline(x = 0.0, color = 'black', linestyle = '-')

for lang_id, (delta_mu_round, delta_mu_unround) in vh_round.items():
    ax_round.scatter(delta_mu_round, delta_mu_unround, color='red')
    ax_round.text(x=delta_mu_round+0.1, y=delta_mu_unround+0.1, s=lang_id, fontsize=14)
for lang_id, (delta_mu_round, delta_mu_unround) in nvh_round.items():
    ax_round.scatter(delta_mu_round, delta_mu_unround, color='green')
    ax_round.text(x=delta_mu_round+0.1, y=delta_mu_unround+0.1, s=lang_id, fontsize=14)
ax_round.set_xlabel("delta_mean_unround", fontsize=20)
ax_round.grid(True, which='both')
# ax_round.set_yticks([])
ax_round.set_ylabel("delta_mean_round", fontsize=20)
ax_round.tick_params(labelsize = 15, axis='both')

ax_round.axhline(y = 0.0, color = 'black', linestyle = '-')
ax_round.axvline(x = 0.0, color = 'black', linestyle = '-')

legend_elements = []
elem = Patch(facecolor='red', edgecolor='black', label="VH Language")
legend_elements.append(elem)
elem = Patch(facecolor='green', edgecolor='black', label="Non-VH Language")
legend_elements.append(elem)
plt.legend(handles=legend_elements, loc='best', bbox_to_anchor=(1.5, 0.5), fontsize=15)

plt.xlim(-4.5, 4.5)
plt.ylim(-4.5, 4.5)

plt.savefig("../img/VH_vs_NVH.png",
            bbox_inches='tight',
            pad_inches=0,
            format='png',
            dpi=300)


plt.show()
```
